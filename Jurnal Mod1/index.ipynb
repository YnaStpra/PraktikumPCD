# 2. Lakukanlah konvolusi pada citra dengan zero padding dan tanpa padding dengan

```python
kernel = np.array([
    [0, 0,-1, 1],
    [0, 2, 1, -1],
    [-1, 1, 2, 0],
    [1, -1, 0, 0]
])
```

```python
result_zero_padding = cv2.filter2D(img, -1, kernel, borderType=cv2.BORDER_CONSTANT)
```

```python
result_no_padding = cv2.filter2D(img, -1, kernel)
```

```python
result_no_clipping = cv2.filter2D(img, -1, kernel, borderType=cv2.BORDER_CONSTANT)
```

```python
result_clipping = cv2.filter2D(img, -1, kernel, borderType=cv2.BORDER_CONSTANT)
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(result_zero_padding, cmap='gray')
plt.title('Result with Zero Padding')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(result_no_padding, cmap='gray')
plt.title('Result without Padding')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(result_clipping, cmap='gray')
plt.title('Result with Clipping')
plt.axis('off')

plt.show()
```
